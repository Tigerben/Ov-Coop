```python
import numpy as np #Importerer bibloteket numpy
import matplotlib.pyplot as plt #Importerer bilboteket pyplot

N = 100 #Merk at N er lengden av matris radene og kolonnene og N^2 er den totale mengden lopper
t = 50000 #Hvor mange "hopp" vi skal se på

NB = np.zeros((N,N)) #Lager en tom "hund" B
NA = NB+1 #Fyller "hund" A med lopper
Andel_lopp = np.zeros(t) #Lager en funksjon av andel lopper på "hund" A

for i in range(0,t):
    loppex = np.random.randint(0,N-1,1) #Finner raden vi skal ta loppe fra
    loppey = np.random.randint(0, N - 1, 1) #Finner loppen vi skal ta fra raden
    if NA[loppex,loppey] == 1: #Sjekker om det er en loppe der
        NB[loppex, loppey] = 1 #Loppen hopper fra A til B om den var i A
        NA[loppex, loppey] = 0
    else: #Dersom loppen ikke er på A
        NB[loppex, loppey] = 0 #Hopper den fra B til A
        NA[loppex, loppey] = 1
    Andel_lopp[i] = 1/N**2*np.sum(NA) #Regner ut andel lopper på "hund" A i tidspunkt i


x = np.arange(0,t,1) #Lager et array med x verdier fra 0 til t

c = 1/N  #Ved å prøve forskjellige verdier finner vi at grafene overlapper best intill likevekts
         #punktet dersom c = 1/N


plt.figure(1) #Lager en pyplot figur
plt.plot(x,Andel_lopp, label="Numeriske lopper") #Plotter andel lopper som en funksjon av tid
plt.plot(x,1/N**2*(N**2/2*(1+np.e**(-2*c**2*x))),label="Analytiske lopper") #Plotter den analytiske andelen lopper
plt.ylim((0,1)) #Tvinger y til å gå fra 0 til 1
plt.legend() #Legger inn labels i figuren
plt.grid() #Legger in et rutenet
plt.show() #Viser funksjonenen


```
